## What is a database?
A database is a way to organise data, such that everything is sorted into rows and columns. 

### Why bother using it?
It can promote **organisation** (keep related data in a logical structure), **efficiency** (be able to quickly find and retrieve specific information), **consistency** (by following specific rules), **scalability** (scale up or down your database) and allow **multi user access**.
